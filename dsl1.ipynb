```python
def display_multiplication_table(num):
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11): # Generates table from 1 to 10
        print(f"{num} x {i} = {num * i}")

# Example usage:
number_to_display = 7
display_multiplication_table(number_to_display)

print("\n") # Add a newline for better separation

number_to_display = 12
display_multiplication_table(number_to_display)
```

5. Develop an Expert System that provides simple decision-making.

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = -3
print(f"The factorial of {number} is {factorial(number)}")
```

## Implement Knowledge Base

### Subtask:
Translate the defined rules into a structured format within Python (e.g., a dictionary, list of tuples, or custom class) that the expert system can interpret. This will serve as the system's knowledge base.


To implement the knowledge base, I will use a list of dictionaries. Each dictionary will represent a rule, containing a list of `conditions` (the 'IF' part) and a `conclusion` (the 'THEN' part). This structure allows for clear representation of each rule and easy iteration when the expert system needs to evaluate them.

**Reasoning**:
Now that the approach is explained, I will implement the knowledge base using a Python list of dictionaries, where each dictionary defines a rule with its conditions and conclusion.



```python
knowledge_base = [
    {
        "conditions": ["symptom_fever", "symptom_cough", "symptom_fatigue"],
        "conclusion": "diagnosis_flu"
    },
    {
        "conditions": ["symptom_runny_nose", "symptom_sore_throat"],
        "conclusion": "diagnosis_common_cold"
    },
    {
        "conditions": ["symptom_stomach_pain", "symptom_nausea"],
        "conclusion": "diagnosis_food_poisoning"
    },
    {
        "conditions": ["symptom_headache", "symptom_sensitivity_to_light"],
        "conclusion": "diagnosis_migraine"
    },
    {
        "conditions": ["symptom_fever", "symptom_rash", "symptom_joint_pain"],
        "conclusion": "diagnosis_dengue_fever"
    }
]

print("Knowledge base created with the following rules:")
for i, rule in enumerate(knowledge_base):
    print(f"Rule {i+1}: IF {', '.join(rule['conditions'])} THEN {rule['conclusion']}")
```

## Develop Inference Engine

### Subtask:
Create a Python function or class that acts as the inference engine. This engine will take user inputs, match them against the rules in the knowledge base, and deduce a conclusion or recommendation.


**Reasoning**:
I will define the `infer_diagnosis` function as described, iterating through the `knowledge_base` and checking if all conditions for any rule are present in the provided symptoms to deduce a conclusion. I will then provide examples to demonstrate its usage.



```python
def infer_diagnosis(observed_symptoms, knowledge_base):
    """
    Inference engine that takes observed symptoms and a knowledge base to deduce a conclusion.

    Args:
        observed_symptoms (list): A list of symptoms observed.
        knowledge_base (list): A list of dictionaries, where each dictionary represents
                               a rule with 'conditions' and a 'conclusion'.

    Returns:
        str: The deduced conclusion or a message indicating no conclusion was found.
    """
    for rule in knowledge_base:
        # Check if all conditions for the current rule are present in the observed symptoms
        all_conditions_met = all(condition in observed_symptoms for condition in rule["conditions"])
        if all_conditions_met:
            return rule["conclusion"]

    return "No diagnosis found based on provided symptoms."

# Example Usage:
# Test case 1: Symptoms matching 'diagnosis_flu'
patient_symptoms_1 = ["symptom_fever", "symptom_cough", "symptom_fatigue", "symptom_sore_throat"]
deduced_diagnosis_1 = infer_diagnosis(patient_symptoms_1, knowledge_base)
print(f"Patient 1 Symptoms: {patient_symptoms_1}")
print(f"Deduced Diagnosis for Patient 1: {deduced_diagnosis_1}\n")

# Test case 2: Symptoms matching 'diagnosis_common_cold'
patient_symptoms_2 = ["symptom_runny_nose", "symptom_sore_throat", "symptom_headache"]
deduced_diagnosis_2 = infer_diagnosis(patient_symptoms_2, knowledge_base)
print(f"Patient 2 Symptoms: {patient_symptoms_2}")
print(f"Deduced Diagnosis for Patient 2: {deduced_diagnosis_2}\n")

# Test case 3: Symptoms matching no specific diagnosis
patient_symptoms_3 = ["symptom_sore_throat", "symptom_tiredness"]
deduced_diagnosis_3 = infer_diagnosis(patient_symptoms_3, knowledge_base)
print(f"Patient 3 Symptoms: {patient_symptoms_3}")
print(f"Deduced Diagnosis for Patient 3: {deduced_diagnosis_3}\n")

# Test case 4: Symptoms matching 'diagnosis_dengue_fever'
patient_symptoms_4 = ["symptom_fever", "symptom_rash", "symptom_joint_pain", "symptom_headache"]
deduced_diagnosis_4 = infer_diagnosis(patient_symptoms_4, knowledge_base)
print(f"Patient 4 Symptoms: {patient_symptoms_4}")
print(f"Deduced Diagnosis for Patient 4: {deduced_diagnosis_4}\n")
```

## Create User Interface

### Subtask:
Implement a simple text-based interface for the user to interact with the expert system. This interface will prompt the user for necessary information and display the system's decision or recommendation.


**Reasoning**:
I will create a Python function named `start_expert_system` that handles user interaction by prompting for symptoms, processing the input, calling the `infer_diagnosis` function, and displaying the result.



```python
def start_expert_system(knowledge_base):
    """
    Implements a simple text-based interface for the expert system.
    """
    print("Welcome to the Medical Diagnosis Expert System!")
    print("This system can help deduce a possible diagnosis based on your symptoms.")

    # Prompt user for symptoms
    user_input = input("\nPlease enter your symptoms, separated by commas (e.g., fever, cough, runny nose): ")

    # Process user input
    # Convert to lowercase, split by commas, and strip whitespace
    observed_symptoms = [symptom.strip().lower() for symptom in user_input.split(',') if symptom.strip()]

    # Map user-friendly symptoms to internal symptom keys for the knowledge base
    # This is a simple mapping. In a real system, a more robust mapping or NLP could be used.
    symptom_mapping = {
        "fever": "symptom_fever",
        "cough": "symptom_cough",
        "fatigue": "symptom_fatigue",
        "runny nose": "symptom_runny_nose",
        "sore throat": "symptom_sore_throat",
        "stomach pain": "symptom_stomach_pain",
        "nausea": "symptom_nausea",
        "headache": "symptom_headache",
        "sensitivity to light": "symptom_sensitivity_to_light",
        "rash": "symptom_rash",
        "joint pain": "symptom_joint_pain"
    }

    mapped_symptoms = []
    for user_symptom in observed_symptoms:
        if user_symptom in symptom_mapping:
            mapped_symptoms.append(symptom_mapping[user_symptom])
        else:
            print(f"Warning: '{user_symptom}' is not a recognized symptom and will be ignored.")


    # Call the inference engine
    diagnosis = infer_diagnosis(mapped_symptoms, knowledge_base)

    # Display the result
    print(f"\nBased on your symptoms, the system suggests: {diagnosis}")

# Example usage of the expert system interface:
start_expert_system(knowledge_base)
```
